## Exercise 5: Calculate Statistics 

### Problem:
Given the array of daily temperatures, calculate:
1. The **mean** (average) temperature
2. The **maximum** temperature
3. The **minimum** temperature
4. The **standard deviation**

```python
temperatures = np.array([72, 75, 78, 82, 79, 73, 76, 80, 77, 74])
```



````python

temperatures = np.array([72, 75, 78, 82, 79, 73, 76, 80, 77, 74])

mean_temp = np.mean(temperatures)
max_temp = np.max(temperatures)
min_temp = np.min(temperatures)
std_temp = np.std(temperatures)

print(f'Mean: {mean_temp}°F')
print(f'Max: {max_temp}°F')
print(f'Min: {min_temp}°F')
print(f'Std Dev: {std_temp:.2f}°F')
````

## Exercise 12: Normalize Data 

### Problem:
**Normalize** the array below:
- Formula: `(x - mean) / std_deviation`




````python

data = np.array([10, 20, 30, 40, 50])

mean = np.mean(data)
std = np.std(data)

normalized = (data - mean) / std

print('Original:', data)
print('Normalized:', normalized)
````

## Exercise 13: Final Challenge - Student Grade Analysis 

### Problem:
Analyze student performance across 3 subjects:




````python

grades = np.array([[85, 90, 78],
                   [92, 88, 95],
                   [78, 85, 80],
                   [90, 92, 88],
                   [88, 86, 91]])




student_avg = np.mean(grades, axis=1)
print('1. Average per student:', student_avg)


subject_avg = np.mean(grades, axis=0)
print('2. Average per subject (Math, English, Science):', subject_avg)


highest_grade = np.max(grades)
print('3. Highest grade overall:', highest_grade)


best_student = np.argmax(student_avg)
print('4. Student with highest average (0-indexed):', best_student)

above_85_count = np.sum(grades > 85)
print('5. Grades above 85:', above_85_count)
````
