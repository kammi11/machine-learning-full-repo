```python
# cleaning the data now...............

# Remove extra spaces and casing.....
df.columns=df.columns.str.strip().str.lower()

df.head()
```

```python
#drop duplicates
df=df.drop_duplicates(keep='first')
df=df.reset_index(drop=True)

df.duplicated().sum()
```

```python
df.head(10)
```

```python
# now code is free from useless columns......

# now fill missing values in many columns......

# drop rows with missing student ids.........
df=df.drop(columns=['student_id'])
df=df.dropna(subset=['exam_score'])
df=df.reset_index(drop=True)
```

```python
df.sample(10)
```

```python
# fill the miss numeriacal values with median values..................
df['age']=df['age'].fillna(df['age'].median())
df['social_media_hours']=df['social_media_hours'].fillna(df['social_media_hours'].median())
df['netflix_hours']=df['netflix_hours'].fillna(df['netflix_hours'].median())
df['attendance_percentage']=df['attendance_percentage'].fillna(df['attendance_percentage'].median())
df['exercise_frequency']=df['exercise_frequency'].fillna(df['exercise_frequency'].median())
df['mental_health_rating']=df['mental_health_rating'].fillna(df['mental_health_rating'].median())
```

```python
# now verify that is there any null values left in these columnssssssssss.
df[['age','social_media_hours','netflix_hours']].isnull().sum()
```

```python
# now fill missing text value(categorical)
df['part_time_job']=df['part_time_job'].fillna(df['part_time_job'].mode()[0])
df['diet_quality']=df['diet_quality'].fillna(df['diet_quality'].mode()[0])
df['internet_quality']=df['internet_quality'].fillna(df['internet_quality'].mode()[0])
df['extracurricular_participation']=df['extracurricular_participation'].fillna(df['extracurricular_participation'].mode()[0])
```

```python
# now preprocessing gender and parental education columns.......
df['gender']=df['gender'].fillna('unknown')
df['parental_education_level']=df['parental_education_level'].fillna('unknown')
```

```python
df.shape
```

```python
# fix spaces in gender 
df['gender'].value_counts()
```

```python
gender_map={
    'female':'Female',
    'male':'Male',
    'f':'Female',
    'm':'Male',
    'mle':'Male',
    'feml':'Female',
    'femle':'Female',
    'F':'Female',
    'M':'Male',
    'FEMALE':'Female',
    'MALE':'Male',
    'unknown':'Unknown',
    'other':'Other'
}
df['gender']=df['gender'].map(gender_map).fillna('Unknown')
df['gender'].value_counts() 
```

```python
# fixing ten and 10 values.............
df['study_hours_per_day']=df['study_hours_per_day'].replace('ten','10')
df['sleep_hours']=df['sleep_hours'].replace('ten','10')
```

```python
# now convert to numeric number type....
df['study_hours_per_day']=pd.to_numeric(df['study_hours_per_day'], errors='coerce')
df['sleep_hours']=pd.to_numeric(df['sleep_hours'], errors='coerce')
```

```python
df['sleep_hours'].value_counts(dropna=False)
```

```python
df.info()
```

```python
# FIX NEGATIVE HOURS - replace negative values with 0
negative_cols = ['study_hours_per_day', 'social_media_hours', 'netflix_hours', 'sleep_hours', 'exercise_frequency']
for col in negative_cols:
    if col in df.columns:
        df[col] = df[col].apply(lambda x: 0 if x < 0 else x)
```

```python
# FIX EXAM_SCORE > 100 - cap at 100
df['exam_score'] = df['exam_score'].apply(lambda x: 100 if x > 100 else x)
```

```python
# Fill remaining missing values in numeric columns with median
df['study_hours_per_day'] = df['study_hours_per_day'].fillna(df['study_hours_per_day'].median())
df['sleep_hours'] = df['sleep_hours'].fillna(df['sleep_hours'].median())
```

```python
# Fill remaining missing values in categorical columns with mode
categorical_cols = ['gender', 'part_time_job', 'diet_quality', 'parental_education_level', 'internet_quality', 'extracurricular_participation']
for col in categorical_cols:
    if col in df.columns and df[col].isnull().sum() > 0:
        df[col] = df[col].fillna(df[col].mode()[0])
```

```python
# Final verification - check for any remaining missing values
print("\nRemaining missing values after all cleaning:")
print(df.isnull().sum())
```

```python
# Final shape
print(f"\nFinal dataset shape: {df.shape}")
```
