```python

balance = 0.0
print("=" * 40)
print("WELCOME TO SIMPLE ATM SYSTEM")
print("=" * 40)
while True:
    print("\n----- ATM MENU -----")
    print("1. Deposit Money")
    print("2. Withdraw Money")
    print("3. View Balance")
    print("4. Exit")
    print("-" * 20)
    choice = input("Enter your choice (1-4): ")
    if choice == "1":
        print("\n--- Deposit Money ---")
        try:
            amount = float(input("Enter amount to deposit: $"))
            
            if amount > 0:
                balance += amount
                print(f"Successfully deposited: ${amount:.2f}")
                print(f"New Balance: ${balance:.2f}")
            else:
                print("Error: Deposit amount must be greater than 0.")
        except ValueError:
            print("Error: Invalid input. Please enter a valid number.")
    elif choice == "2":
        print("\n--- Withdraw Money ---")
        try:
            amount = float(input("Enter amount to withdraw: $"))
            
            if amount > 0:
                if amount <= balance:
                    balance -= amount
                    print(f"Successfully withdrew: ${amount:.2f}")
                    print(f"Remaining Balance: ${balance:.2f}")
                else:
                    print(f"Error: Insufficient balance. Your current balance is: ${balance:.2f}")
            else:
                print("Error: Withdrawal amount must be greater than 0.")
        except ValueError:
            print("Error: Invalid input. Please enter a valid number.")
    elif choice == "3":
        print("\n--- Current Balance ---")
        print(f"Your current balance is: ${balance:.2f}")
    elif choice == "4":
        print("\nThank you for using our ATM system!")
        print("Have a great day!")
        break
    else:
        print("\nError: Invalid choice. Please select a valid option (1-4).")
```

```python


def read_emails_from_file(filename):
    emails = []
    try:
        with open(filename, 'r') as file:
            for line in file:
                email = line.strip()
                if email:
                    emails.append(email)
    except FileNotFoundError:
        print(f"Error: {filename} not found.")
        return []
    return emails

def classify_email(email):
    email_lower = email.lower()
    spam_keywords = ['free', 'winner', 'congratulations', 'lottery', 'prize', 
                     'urgent', 'click here', 'limited time', 'offer']
    for keyword in spam_keywords:
        if keyword in email_lower:
            return "Spam"
    return "Not Spam"

def display_all_emails(emails):
    print("\n" + "="*60)
    print("ALL EMAILS WITH THEIR LABELS")
    print("="*60)
    
    for i, email in enumerate(emails, 1):
        label = classify_email(email)
        print(f"\nEmail {i}:")
        print(f"Content: {email}")
        print(f"Label: {label}")
        print("-"*40)

def display_spam_emails(emails):
    print("\n" + "="*60)
    print("SPAM EMAILS ONLY")
    print("="*60)
    
    spam_found = False
    for i, email in enumerate(emails, 1):
        label = classify_email(email)
        if label == "Spam":
            spam_found = True
            print(f"\nSpam Email {i}:")
            print(f"Content: {email}")
            print("-"*40)
    
    if not spam_found:
        print("\nNo spam emails found.")

def main():
    print("SPAM EMAIL DETECTION SYSTEM")
    print("="*60)
    
    filename = "emails"
    emails = read_emails_from_file(filename)
    
    if not emails:
        print("No emails to process. Please check the file.")
        return
    
    while True:
        print("\n" + "="*60)
        print("MENU")
        print("="*60)
        print("1. Display all emails with labels")
        print("2. Display only spam emails")
        print("3. Exit")
        
        choice = input("\nEnter your choice (1-3): ")
        
        if choice == '1':
            display_all_emails(emails)
        elif choice == '2':
            display_spam_emails(emails)
        elif choice == '3':
            print("\nThank you for using Spam Email Detection System. Goodbye!")
            break
        else:
            print("\nInvalid choice. Please enter 1, 2, or 3.")

if __name__ == "__main__":
    main()
```
