```python
def count_words(sentence):
    words_dict ={}
    word = sentence.split()
    for i in word:
        if i in words_dict:
            words_dict[i] += 1
        else:
            words_dict[i] = 1
    return words_dict
    
```

```python
import numpy as np 

price = np.array([85,120,95,150,200,75])

#calculate the mean and median price
mean_price = np.mean(price)
median_price = np.median(price)
print("Mean price: ", mean_price)
#median is better than mean 
print("median price:", median_price)

filtered_price = price[price > 100]
print("prices > 100:", filtered_price)
```
